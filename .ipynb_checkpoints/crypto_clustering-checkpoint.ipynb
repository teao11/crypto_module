# Clustering Crypto

```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
data_df = pd.read_csv("crypto_data.csv")
```

```python
# Keep all the cryptocurrencies that are being traded.
data_df = data_df[data_df["IsTrading"] == True]
data_df
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
data_df = data_df[(data_df["Algorithm"]!="N/A")]
```

```python
# Remove the "IsTrading" column. 
data_df.drop(columns=["IsTrading"], inplace=True)
```

```python
# Remove rows that have at least 1 null value.
```

```python
# Keep the rows where coins are mined.
data_df = data_df[(data_df["TotalCoinsMined"]>0)]
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
coin_names = pd.DataFrame(data_df["CoinName"])
coin_names
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
data_df.drop(columns=["CoinName"],inplace=True)
```

```python
# Use get_dummies() to create variables for text features.
X=pd.get_dummies(data=data_df)
```

```python
# Standardize the data with StandardScaler().
X= StandardScaler().fit_transform(X)
print(X[:10])
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
pca = PCA(n_components=3)
pca_components = pca.fit_transform(X)
pca_components
```

```python
# Create a DataFrame with the three principal components.
unsupervised_df = pd.DataFrame(data=pca_components,
                               columns=["PC 1", "PC 2", "PC 3"],
                               index=data_df.index)

unsupervised_df.head()
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
inertia = []
elbow = list(range(1, 11))

# Inertia for K values
for i in elbow:
    kmeans = KMeans(n_clusters=i, random_state=21)
    kmeans.fit(pca_components)
    inertia.append(kmeans.inertia_)

# Elbow curve generated
elbow_data = {"k": elbow, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=elbow, title="Elbow Curve")
```

Running K-Means with `k=4`

```python
# Initialize the model.
uns_model = KMeans(n_clusters=4, random_state=4)

# Fit the model to the PCA
uns_model.fit(pca_components)

# Predict clusters using model from above
pred = uns_model.predict(pca_components)
pred
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
cluster_df = pd.concat([data_df, unsupervised_df], axis=1)

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
cluster_df["CoinName"] = coin_names["CoinName"]

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
cluster_df["Class"] = uns_model.labels_
cluster_df.head()

# Print the shape of the clustered_df
print(cluster_df.shape)
cluster_df.head(10)
```

### Deliverable 4: Visualizing Cryptocurrencies Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(cluster_df,
                    x="PC 1",
                    y="PC 2",
                    z="PC 3",
                    color="Class",
                    symbol="Class",
                    width=800,
                    # hover_name=["CoinName"],
                    # hover_data=["Algorithm"] HAVING ISSUES ):
                   )

fig.update_layout(
    legend=dict(x=0, y=1)
    )

fig.show()
```
